```python
#writing file names and data frame names
year=list(range(2008,2024,1))
file_name=list('ForbesGlobal2000_'+str(x) for x in year)
df_name=list('df_'+str(x) for x in year)
len(year)
```

```python
#collecting my data frames in a dictionary
data_frames={}
for i in range(len(df_name)):
  data_frames[df_name[i]] = pd.read_excel(file_name[i]+'.xlsx')
```

```python
#Meriging all my dataframes by company, industry and country
merged_df = None  # Initialize the merged dataframe

for df_key, df_value in data_frames.items():
    if merged_df is None:
        merged_df = df_value.copy()  # Initialize merged_df with the first dataframe
    else:
        merged_df = pd.merge(merged_df, df_value, on=['Company','Industry','Country'], how='outer')
```

```python
merged_df_cleaned=merged_df.dropna() #returns companies tha have been every year in the ranking
merged_df_cleaned.shape
```

```python
#dataset of the ranking by year
rank_df_columns=list('Rank_nr_'+str(x) for x in year)
my_list=['Company','Industry','Country']+rank_df_columns
ranking_num=merged_df.loc[:,my_list]
ranking_num.head()
```
